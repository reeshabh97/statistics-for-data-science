```python
# Code for calculating NORMAL or GAUSSIAN probability for each observation:

import numpy as np

# Creating a Function.
def normal_dist(x, mean, sd):
    prob_density = (1/(sd*np.sqrt(2*np.pi)))*(np.exp(-0.5*((x-mean)/sd)**2))
    return prob_density
 
# Apply function to the data.
# Estimated fractions using normal distribution.
max=0
for x in range(50,351,1):
    pdf = normal_dist(x,185,50)
    if pdf>max:
        max=pdf
    print(pdf)
```
